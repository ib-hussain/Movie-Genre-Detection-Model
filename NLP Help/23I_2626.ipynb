2(iii. Generate Poem

Use the unigram, bigram and trigram models to predict the next word, continuing until the verse is complete. Selecting subsequent words after choosing the first word requires predicting the most probable next word. Utilize Conditional Frequency Distribution (CFD) to determine the set of words that occur most frequently after the selected
Apply the same approach to generate 3 verses of 4 lines each (7-10 words per line).

```python
import nltk
from nltk.util import ngrams
from nltk.probability import ConditionalFreqDist
import random

words = sheikh
#Generate unigrams,bigrams, trigrams
unigrams = words
bigrams = list(ngrams(words, 2))
trigrams = list(ngrams(words, 3))

unigram_cfd = ConditionalFreqDist((w, w) for w in unigrams)
bigram_cfd = ConditionalFreqDist((w1, w2) for w1, w2 in bigrams)
trigram_cfd = ConditionalFreqDist(((w1, w2), w3) for w1, w2, w3 in trigrams)

def generate_line(start_word, model_cfd, n=7):
    line = [start_word]
    while len(line) < n:
        last_word = line[-1] if len(line) == 1 else tuple(line[-2:])
        if len(line) >= 2 and last_word in model_cfd:
            next_word = model_cfd[last_word].max()
        elif last_word in bigram_cfd:
            next_word = bigram_cfd[last_word].max()
        else:
            next_word = random.choice(unigram_cfd.keys())
        line.append(next_word)
    return ' '.join(line)

def generate_stanza(starting_words, n_lines=4, line_length=(7, 10)):
    stanza = []
    for _ in range(n_lines):
        start_word = random.choice(starting_words)
        line_length_random = random.randint(*line_length)
        stanza.append(generate_line(start_word, trigram_cfd, n=line_length_random))
    return stanza

def generate_poem(starting_words, n_stanzas=3):
    poem = []
    for _ in range(n_stanzas):
        stanza = generate_stanza(starting_words)
        poem.append(stanza)
    return poem

poem = generate_poem(starting_words)

for stanza in poem:
    for line in stanza:
        print(line)
    print()

```
